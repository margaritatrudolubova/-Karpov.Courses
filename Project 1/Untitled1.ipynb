**6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации.**
**Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.**
**В кластеризации можешь выбрать следующие метрики:** 
**R - время от последней покупки пользователя до текущей даты,** 
**F - суммарное количество покупок у пользователя за всё время,** 
**M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры.**
**Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров.**
**Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.** 

```python
#создадим датафрейм для работы с данными

df_12= df_1.merge(df_2, how='inner', on='customer_id')
df_rfm= df_12.merge(df_3, how='inner', on='order_id')

df_rfm.head()
```

```python
print('{:,} invoices don\'t have a customer id'.format(df_rfm[df_rfm.customer_id.isnull()].shape[0]))
```

```python
#Найдем временные рамки наших данных
print('Orders from {} to {}'.format(df_rfm['order_purchase_timestamp'].min(),
                                    df_rfm['order_purchase_timestamp'].max()))
```

```python
#у нас есть данные о продаж за 2 года. Возьмем этот период для анализа RFM. Я установлю период исследования на 730 дней.
```

```python
#посмотрим на количество строк в каждом счете
df_rfm['order_id'].value_counts().head()
```

```python
#создадим датафрейм данных, который будет агрегировать наши продажи на уровне заказа.
orders =df_rfm.groupby(['order_id', 'order_purchase_timestamp', 'customer_id']).agg({'price': lambda x: x.sum()}).reset_index()
orders.head()
```

```python
#смоделируем анализ, установив NOW дату на один день после последней покупки. Эта дата будет использоваться в качестве ориентира для подсчета Recency баллов.
```

```python
NOW = orders['order_purchase_timestamp'].max() + timedelta(days=1)
NOW
```

```python
period = 730
```

```python
#Рассчитаем недавность, частоту и денежную стоимость каждого клиента
```

```python
#Чтобы упростить задачу, добавим столбец с количеством дней между покупкой и настоящим моментом. Чтобы найти значения недавности, нужно найти минимум этого столбца для каждого клиента.
```

```python
orders['DaysSinceOrder'] = orders['order_purchase_timestamp'].apply(lambda x: (NOW - x).days)
```

```python
#Баллы рассчитываются для каждого клиента. Результаты будут храниться в столбцах.
```

```python
aggr = {
    'DaysSinceOrder': lambda x: x.min(),  # the number of days since last order (Recency)
    'order_purchase_timestamp': lambda x: len([d for d in x if d >= NOW - timedelta(days=period)]), # the total number of orders in the last period (Frequency)
}
rfm = orders.groupby('customer_id').agg(aggr).reset_index()
rfm.rename(columns={'DaysSinceOrder': 'Recency', 'order_purchase_timestamp': 'Frequency'}, inplace=True)

rfm.head()
```

```python
#У нас есть данные о недавности и частоте. Теперь добавим денежную стоимость каждого клиента, добавив продажи за последний год.
rfm['Monetary'] = rfm['customer_id'].apply(lambda x: orders[(orders['customer_id'] == x) & \
                                                           (orders['order_purchase_timestamp'] >= NOW - timedelta(days=period))]\
                                                           ['price'].sum())
rfm.head()
```

```python
#Рассчитываем баллы R, F и M
```

```python
#используем метод квинтилей
quintiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([.4, .6, .8]).to_dict()
quintiles
```

```python
#Напишем методы для присвоения рангов от 1 до 3. Чем меньше значение Недавности, тем лучше, когда более высокие значения Частоты и Денег лучше. 
```

```python
def r_score(x):
    if   x <= quintiles['Recency'][.4]:
        return 3
    elif x <= quintiles['Recency'][.6]:
        return 2
    else:
        return 1
```

```python
def fm_score(x, c):
    if x <= quintiles[c][.4]:
        return 1
    elif x <= quintiles[c][.6]:
        return 2
    else:
        return 3
```

```python
# Рассчитаем оценки R, F и M каждого клиента.
```

```python
rfm['R'] = rfm['Recency'].apply(lambda x: r_score(x))
```

```python
rfm['F'] = rfm['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
```

```python
rfm['M'] = rfm['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))
```

```python
#Объединим все оценки в новую колонку
rfm['RFM_score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)
rfm.head()
```

```python
# У нас достаточно много получается сегментов, но мы будем работать только с самыми важными
```

```python
segt_map = {
    r'[1][1]': 'in hibernation',
    r'[1][2]': 'rare_customers',
    r'[1]3': 'long_time',
    r'2[1]': 'about to sleep',
    r'22': 'need attention',
    r'[2][3]': 'loyal customers',
    r'31': 'promising',
    r'32': 'potential loyalists',
    r'[3][3]': 'champions'
}

rfm['Segment'] = rfm['R'].map(str) + rfm['F'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)

rfm.head()
```

```python
#посмотрим на визуализацию распределения клиентов по R и F
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for i, p in enumerate(['R', 'F']):
    parameters = {'R':'Recency', 'F':'Frequency'}
    y = rfm[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Distribution of {}'.format(parameters[p]),
                fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('firebrick')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 5,
                '{}\n({}%)'.format(int(value), int(value * 100 / y.sum())),
               ha='center',
               va='top',
               color='w')

plt.show()
```

```python
#посмотрим на гистограмму распределения клиентов
```

```python
segments_counts = rfm['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
              segments_counts,
              color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
        value = bar.get_width()
        if segments_counts.index[i] in ['champions', 'loyal customers']:
            bar.set_color('firebrick')
        ax.text(value,
                bar.get_y() + bar.get_height()/2,
                '{:,} ({:}%)'.format(int(value),
                                   int(value*100/segments_counts.sum())),
                va='center',
                ha='left'
               )

plt.show()
```
