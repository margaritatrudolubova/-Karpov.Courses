**В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?**

**Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?**

**Посмотрим какие метрики можно отследить при 1 исходе событий:**

Для начала стоит посмотреть какое количество пользователей заинтересовались нашим событием и стали заходить в приложение:
1. Метрика вовлеченности (Engagement) = (TE/TF)*100%, где
    TE(Total Engagement) - общая вовлеченность клиентов(для нас это кол-во пользователей вошедших в приложение на игру)
    а TF(Total Followers)- общее количество пользователей (число людей, которые зарегестрированы в нашей игре).
***    
2. Далее обратим внимание на метрику:
    Monthly active users = Количество активных пользователей в месяц
Посмотрим как сильно падает данная метрика, если сильно, то можно попробовать отправлять пуш-уведомления для ее повышения.
***
3. Session duration= Длительность сессии. 
Если длительность сессии будет небольшой, то можно обратить внимание на сложность уровней, может действительно стоит сделать их чуточку сложнее или легче.
***
4. Заодно посмотрим на метрику: 
    ASD = Суммарная продолжительность сессий за период / общее количество сессий за тот же период. 
При сравнении разных групп пользователей по ASD может оказаться, что одни запускали игру каждый день, но всего на пару минут. Другие же проводили в ней по три часа, но только раз в неделю.
***
5. Winners (победители) = количество пользователей, успешно прошедших событие / EAU(event active users). 
Посмотрим, сколько пользователей дошли до конца нашего события и стали победителями предметов, бонусов и тд. Если показатели низкие, то можно пересмотреть количество уровней и сделать их чуть полегче, если показатели высокие, то можно наоборот увеличить количество уровней и их сложность.
***
6. Посмотрим на средний чек в игре. 
    ATV(Average Transaction Value) = Общая сумма покупок / количество покупок.
Average Transaction Value — средняя стоимость покупок за период, средний чек. Если этот показатель высокий, то очень замечательно, а если низкий, то можно расширить ассортимент или предложить дополнительные продукты на этапе покупки. Специальные предложения тоже мотивируют пользователей покупать больше.
***
7. Neutral (нейтральные пользователи) = EAU/ общее количество пользователей. 
Посмотрим сколько пользователей зашло на событие, но никак не отреагировали на него (не стали проходить), в причинах такого решения пользователя стоит разбираться более детально.
***
8. LTV = суммарный доход, полученный с новых пользователей за промежуток времени нашего события. 
Чтобы повысить LTV, нужно постоянно поддерживать вовлеченность пользователей — мониторить их отзывы, добавлять новые функции, инструменты, бонусы.


**Теперь посмотрим на исход событий при откатывании пользователя назад на несколько уровней:**

Рассмотрев более подробно данное событие, можно обойтись метриками, описанными выше, но стоит учесть, что некоторые показатели будут намного меньше, чем при обычном прохождении игры. Не все готовы идти до конца, постоянно усложняя себе задачу за проигрыш. 

Один из вариантов проверить два исхода событий, разделить пользователей на 2 группы или сделать небольшие выборки и провести A/B-тестирование, где для 1 группы будут обычные уровни без откатывания назад, а для 2 группы с откатыванием. 

В дальнейшем можно будет проанализировать каждый уровень отдельно, чтобы понять на каком именно уровне отваливается большинство пользователей и как вариант, либо упростить уровень, либо мотивировать какими-то бонусами, к примеру, дополнительная жизнь для прохождения уровня.
